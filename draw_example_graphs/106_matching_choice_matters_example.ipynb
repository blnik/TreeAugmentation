```python
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
def draw_edge(tree, edge, color, curvature, edge_label = '', edge_width = 2, linestyle = '--'):
    nx.draw_networkx_edges(
        tree, 
        pos = pos, 
        edgelist = [edge],  
        edge_color = color,
        connectionstyle=f'arc3,rad={curvature}', 
        arrowstyle = '-',
        style = linestyle,
        width = edge_width
    )
    nx.draw_networkx_edge_labels(tree, pos = pos, edge_labels = {edge: edge_label}, font_size = 12, font_color = 'C0', bbox = dict(alpha=0))
```

```python
fig, ax = plt.subplots(1,1,tight_layout = True)
fig.set_size_inches(8, 4)

plt.sca(ax)
plt.axis('off')
tree = nx.Graph([(2,1), (1,3), (1,4), (5,4), (6,4), (7,5), (8,5), (9,5)])
links = [(3,5), (1,5), (0,2), (2,7), (0,3), (3,7)]

directed_tree = nx.bfs_tree(tree, 1)
pos = graphviz_layout(directed_tree, prog="dot")
nx.draw_networkx(
    directed_tree,
    pos = pos,
    edge_color = 'black',
    node_color = 'silver',
    arrowstyle = '-',
    width = 3,
    with_labels = True
)


draw_edge(directed_tree, (6,9), 'C0', 0.2)
draw_edge(directed_tree, (7,8), 'C0', 0.2)
draw_edge(directed_tree, (8,3), 'C0', -0.2)
draw_edge(directed_tree, (7,2), 'C0', -0.2)
draw_edge(directed_tree, (2,3), 'C0', 0.2)


plt.savefig('../report_images/tree_examples/matching_choice_matters_example.png', dpi = 200)
plt.show()
```
