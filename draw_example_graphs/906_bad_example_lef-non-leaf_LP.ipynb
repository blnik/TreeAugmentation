```python
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
def draw_edge(tree, edge, color, curvature, edge_label = '', label_pos = 0.5):
    nx.draw_networkx_edges(
        tree, 
        pos = pos, 
        edgelist = [edge],  
        edge_color = color,
        connectionstyle=f'arc3,rad={curvature}', 
        arrowstyle = '-',
        style = '--',
        width = 1.5
    )
    nx.draw_networkx_edge_labels(tree, pos = pos, edge_labels = {edge: edge_label}, font_color = color, bbox = dict(alpha=0), label_pos = label_pos)
```

```python
tree = nx.DiGraph([
    (1,3), (2,4), (3,5), (4,5), (5,6), (6,7), (7,8), (7,9), (8,10), (8,11), (11,13), (11,12), (13,14), (14,15), 
    (14,16), (15,17), (16,18)
])

pos = {
    1: (0,1), 
    2: (0,0), 
    3: (1,1),
    4: (1,0),
    5: (2,0.5),
    6: (3,0.5),
    7: (4,0.5),
    8: (5,1),
    9: (5,0),
    10: (6,1.5),
    11: (6,0.5),
    13: (8,1),
    12: (7,0),
    14: (9,1),
    15: (10,1.5),
    16: (10,0.5),
    17: (11,1.5),
    18: (11,0.5)
}

fig, ax = plt.subplots(1,1,tight_layout = True)
fig.set_size_inches(12, 4)

plt.sca(ax)
plt.axis('off')
#plt.axis('equal')

nx.draw_networkx(tree, pos,
    edge_color = 'black', #black
    node_color = 'black', #silver/black
    node_size = 100, # None
    arrowstyle = '-',
    width = 1.5, # 3
    with_labels = False # True
)


nx.draw_networkx_edge_labels(tree, pos = pos, edge_labels = {
    (6,7): 'a',
    (7,9): 'b',
    (7,8): 'c',
    (8,10): 'd',
    (8,11): 'e',
    (11,12): 'g',
    (11,13): 'f'
})

c_1 = 0.4
c_0 = 0.2

draw_edge(tree, (1,4), 'C0', -c_0)
draw_edge(tree, (2,3), 'C0', c_0)
draw_edge(tree, (1,6), 'C0', -c_1)
draw_edge(tree, (2,6), 'C0', c_1)

draw_edge(tree, (6,9), 'C0', 0.2)
draw_edge(tree, (9,11), 'C0', 0.2)
draw_edge(tree, (12,8), 'C0', -0.3)
draw_edge(tree, (12,13), 'C0', 0.1)
draw_edge(tree, (10,7), 'C0', 0.3)
draw_edge(tree, (10,13), 'C0', -0.1)

draw_edge(tree, (17,16), 'C0', c_0)
draw_edge(tree, (18,15), 'C0', -c_0)
draw_edge(tree, (17,13), 'C0', c_1)
draw_edge(tree, (18,13), 'C0', -c_1)

ht = -0.4
plt.plot([0,3], [ht,ht], color = 'k', marker = '|')
plt.plot([3,8], [ht,ht], color = 'k', marker = '|')
plt.plot([8,11], [ht,ht], color = 'k', marker = '|')
plt.text(1.5, ht-0.1, 'A', horizontalalignment = 'center', verticalalignment = 'center', color = 'k')
plt.text(5.5, ht-0.1, 'B', horizontalalignment = 'center', verticalalignment = 'center', color = 'k')
plt.text(9.5, ht-0.1, 'C', horizontalalignment = 'center', verticalalignment = 'center', color = 'k')

plt.text(4.3,0,'1', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(5.3,0,'2', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(6.6,0,'3', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(7.4,0.2,'4', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(5.3,1.45,'5', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(6.6,1.45,'6', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')

plt.savefig('../report_images/tree_examples/bad_example_leaf-non-leaf_LP.png', dpi = 200)
plt.show()
```
