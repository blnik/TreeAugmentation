```python
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
def draw_edge(tree, edge, color, curvature, edge_label):
    nx.draw_networkx_edges(
        tree, 
        pos = pos, 
        edgelist = [edge],  
        edge_color = color,
        connectionstyle=f'arc3,rad={curvature}', 
        arrowstyle = '-',
        style = '--',
        width = 1.5
    )
    nx.draw_networkx_edge_labels(tree, pos = pos, edge_labels = {edge: edge_label}, font_color = 'C0', bbox = dict(alpha=0))
```

```python
fig, ax = plt.subplots(1,1,tight_layout = True)
fig.set_size_inches(12, 4)

plt.sca(ax)
plt.axis('off')
tree = nx.Graph([
    (1,2), (1,3), (2,4), (2,5), (3,6), (3,7), (4,8), (4,9), (5,10), (5,11), (6,12), (6,13), (7,14), (7,15),
    (8,16), (8,17), (9,18), (9,19), (10,20), (10,21), (11,22), (11,23), (12,24), (12,25), (13,26), (13,27),
    (14,28), (14,29), (15,30), (15,31)
])

directed_tree = nx.bfs_tree(tree, 1)
pos = graphviz_layout(directed_tree, prog="dot")
nx.draw_networkx(
    directed_tree,
    pos = pos,
    edge_color = 'black',
    node_color = 'black',
    node_size = 100,
    arrowstyle = '-',
    width = 1.5,
    with_labels = False
)

draw_edge(directed_tree, (16,17), 'C0', 0.5, 1)
draw_edge(directed_tree, (17,18), 'C0', 0.5, 2)
draw_edge(directed_tree, (18,19), 'C0', 0.5, 3)
draw_edge(directed_tree, (19,20), 'C0', 0.5, 4)
draw_edge(directed_tree, (20,21), 'C0', 0.5, 5)
draw_edge(directed_tree, (21,22), 'C0', 0.5, 6)
draw_edge(directed_tree, (22,23), 'C0', 0.5, 7)
draw_edge(directed_tree, (23,24), 'C0', 0.5, 8)
draw_edge(directed_tree, (24,25), 'C0', 0.5, 9)
draw_edge(directed_tree, (25,26), 'C0', 0.5, 10)
draw_edge(directed_tree, (26,27), 'C0', 0.5, 11)
draw_edge(directed_tree, (27,28), 'C0', 0.5, 12)
draw_edge(directed_tree, (28,29), 'C0', 0.5, 13)
draw_edge(directed_tree, (29,30), 'C0', 0.5, 14)
draw_edge(directed_tree, (30,31), 'C0', 0.5, 15)


nx.draw_networkx_edge_labels(
    directed_tree, pos = pos, 
    edge_labels = {
        (4,9): 'a',
        (5,10): 'b', 
        (5,11): 'c',
        (6,12): 'd',
        (6,13): 'e',
        (7,14): 'f'
    }, font_color = 'k', rotate = False
)


plt.savefig('../report_images/tree_examples/badLink_example.png', dpi = 200)
plt.show()
```
