```python
# working with graphs
import networkx as nx

# plotting 
import matplotlib.pyplot as plt
%matplotlib inline
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
# links example
n = 10
tree = nx.random_tree(n, seed = 2)
root = 5

directed_tree = nx.bfs_tree(tree, root)
pos = graphviz_layout(directed_tree, prog="dot")
```

```python
def draw_edge(edge, color, curvature):
    nx.draw_networkx_edges(
        directed_tree, 
        pos = pos, 
        edgelist = [edge],  
        edge_color = color,
        connectionstyle=f'arc3,rad={curvature}', 
        arrowstyle = '-',
        style = '--',
        width = 2
    )
```

```python
def draw_tree():
    nx.draw_networkx(
        directed_tree,
        pos = pos,
        edge_color = 'black',
        node_color = 'silver',
        arrowstyle = '-',
        width = 3,
        with_labels = False
    )
    
    labels = {}

    k = 1
    for node in directed_tree.nodes():
        if node == root:
            labels[root] = '$r$'
        else:
            labels[node] = k
            k += 1

    nx.draw_networkx_labels(directed_tree,pos,labels)
```

```python
fig, ax = plt.subplots(tight_layout = True)
fig.set_size_inches(6, 4)


plt.sca(ax)
plt.axis('off')

# plot the original tree
draw_tree()



# plot the links
draw_edge((2,9), 'C0', -0.2)
draw_edge((3,8), 'C1', 0.2)
draw_edge((0,6), 'C2', 0.2)
draw_edge((root, 8), 'C0', -0.9)
plt.axis('off')

plt.savefig('../presentation_images/link_types_example.png', dpi = 200)
plt.show()
```
