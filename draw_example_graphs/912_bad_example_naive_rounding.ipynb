```python
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
def draw_edge(tree, edge, color, curvature, edge_label):
    nx.draw_networkx_edges(
        tree, 
        pos = pos, 
        edgelist = [edge],  
        edge_color = color,
        connectionstyle=f'arc3,rad={curvature}', 
        arrowstyle = '-',
        style = '--',
        width = 1.5
    )
    nx.draw_networkx_edge_labels(tree, pos = pos, edge_labels = {edge: edge_label}, font_color = 'C0', bbox = dict(alpha=0))
```

```python
fig, ax = plt.subplots(1,1,tight_layout = True)
fig.set_size_inches(8, 4)

plt.sca(ax)
plt.axis('off')
edgeset = [(1,3), (1,4), (1,2), (1,7), (1,8), (2,5), (2,6)]
tree = nx.Graph(edgeset)
directed_tree = nx.DiGraph(edgeset)

pos = {
    3: (0,0),
    4: (1,0),
    5: (2,0),
    6: (3,0),
    7: (4,0),
    8: (5,0),
    2 : (2.5,1),
    1: (2.5,2)
}

nx.draw_networkx(
    tree,
    pos = pos,
    edge_color = 'black',
    node_color = 'black',
    node_size = 100,
    arrowstyle = '-',
    width = 1.5,
    with_labels = False
)

draw_edge(directed_tree, (3,4), 'C0', 0.3,'')
draw_edge(directed_tree, (4,5), 'C0', 0.3,'')
draw_edge(directed_tree, (5,6), 'C0', 0.3,'')
draw_edge(directed_tree, (6,7), 'C0', 0.3,'')
draw_edge(directed_tree, (7,8), 'C0', 0.3,'')
draw_edge(directed_tree, (3,5), 'C0', 0.5,'')
draw_edge(directed_tree, (6,8), 'C0', 0.5,'')
draw_edge(directed_tree, (4,7), 'C0', 0.36,'')

plt.text(0.5, -0.05, '1', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(1.5, -0.05, '3', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(2.5, -0.05, '4', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(3.5, -0.05, '5', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(4.5, -0.05, '7', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(1, -0.45, '2', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(4, -0.45, '6', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')
plt.text(2.5, -0.55, '8', horizontalalignment = 'center', verticalalignment = 'center', color = 'C0')

plt.savefig('../report_images/tree_examples/naive_rounding_bad_example.png', dpi = 200)
plt.show()
```
