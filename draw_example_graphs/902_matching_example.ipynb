```python
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
def draw_edge(tree, edge, color, curvature):
    nx.draw_networkx_edges(
        tree, 
        pos = pos, 
        edgelist = [edge],  
        edge_color = color,
        connectionstyle=f'arc3,rad={curvature}', 
        arrowstyle = '-',
        style = '--',
        width = 2
    )
```

```python
fig, ax = plt.subplots(1,2,tight_layout = True)
fig.set_size_inches(12, 4)

plt.sca(ax[0])
plt.axis('off')
tree = nx.Graph([(1,2), (2,3), (3,4), (3,8), (4,5), (4,6), (4,7), (8,9), (9,10), (9,11)])

directed_tree = nx.bfs_tree(tree, 2)
pos = graphviz_layout(directed_tree, prog="dot")
nx.draw_networkx(
    directed_tree,
    pos = pos,
    edge_color = 'black',
    node_color = 'silver',
    arrowstyle = '-',
    width = 3,
    with_labels = True
)
draw_edge(directed_tree, (1,5), 'C1', 0.3)
draw_edge(directed_tree, (5,6), 'C0', 0.3)
draw_edge(directed_tree, (7,10), 'C0', 0.3)
draw_edge(directed_tree, (10,11), 'C1', 0.3)


plt.sca(ax[1])
plt.axis('off')
tree = nx.Graph([('1-5', 6), ('1-5', 7), ('1-5', 8), (8, '9-11')])
directed_tree = nx.bfs_tree(tree, '1-5')
pos = graphviz_layout(directed_tree, prog="dot")

nx.draw_networkx_nodes(
    directed_tree,
    pos = pos,
    node_color = 'silver',
    node_shape = 's',
    nodelist = ['1-5', '9-11'],
    node_size = 900
)

nx.draw_networkx_nodes(
    directed_tree,
    pos = pos,
    node_color = 'silver',
    nodelist = [6, 7, 8],
)

nx.draw_networkx_edges(
    directed_tree,
    pos = pos,
    edge_color = 'black',
    arrowstyle = '-',
    width = 3
)

nx.draw_networkx_labels(
    directed_tree,
    pos = pos
)
draw_edge(directed_tree, (6,'1-5'), 'C0', 0.3)
draw_edge(directed_tree, (7,'9-11'), 'C0', 0.3)


plt.savefig('../report_images/tree_examples/matching_example.png', dpi = 200)
plt.show()
```
