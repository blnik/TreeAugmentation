## Find Maximum Approximation Ratio for Matching Heuristic

```python
samples = 50000
n = 10
factor = 2
ar_rg = mat.get_approximation_ratios_of_matchingHeuristic_approximation(
    'iterative_removal', 'greedy', samples, n, factor, 'saved_outputs'
)

ar_rm = mat.get_approximation_ratios_of_matchingHeuristic_approximation(
    'iterative_removal', 'leaf_distance_matching', samples, n, factor, 'saved_outputs'
)

ar_wg = mat.get_approximation_ratios_of_matchingHeuristic_approximation(
    'max_weight', 'greedy', samples, n, factor, 'saved_outputs'
)

ar_wm = mat.get_approximation_ratios_of_matchingHeuristic_approximation(
    'max_weight', 'leaf_distance_matching', samples, n, factor, 'saved_outputs'
)

print('maximum approximation ratio for mat RG:', np.nanmax(ar_rg))
print('maximum approximation ratio for mat WG:', np.nanmax(ar_wg))
print('maximum approximation ratio for mat RM:', np.nanmax(ar_rm))
print('maximum approximation ratio for mat WM:', np.nanmax(ar_wm))
```
